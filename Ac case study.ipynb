```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the data
ac=pd.read_csv(r"C:\Users\manis\Downloads\AC_Data.csv")
```

```python
ac.head()
```

```python
#number of null values
ac.isnull().sum()
```

```python
#dropping all the null values present 
ac=ac.dropna()
```

```python
100*(ac.isnull().sum()/len(ac))
```

```python
ac.rename(columns={'0':'Timestamp'},inplace= True)
```

```python
ac.reset_index(inplace=True)
```

```python
ac.drop('index',axis=1,inplace=True)
```

```python
ac.head()
```

```python
#making new column from the timestamp column
ac['Year']=pd.DatetimeIndex(ac['Timestamp']).year
ac['Month']=pd.DatetimeIndex(ac['Timestamp']).month
ac['Day']=pd.DatetimeIndex(ac['Timestamp']).day
ac['Hour']=pd.DatetimeIndex(ac['Timestamp']).hour
ac['Minute']=pd.DatetimeIndex(ac['Timestamp']).minute
ac['Weekday']=pd.DatetimeIndex(ac['Timestamp']).weekday
```

```python
ac.head()
```

```python
mnth=[]
for i in ac['Month']:
    if i==8:
        mnth.append('August')
    else:
        mnth.append('September')
        
ac['month']=mnth 
```

```python
ac.Weekday.value_counts()
```

```python
week=[]
for i in ac['Weekday']:
    if i==0:
        week.append('Monday')
    if i==1:
        week.append('Tuesday')
    if i==2:
        week.append('Wednesday')
    if i==3:
        week.append('Thursday')
    if i==4:
        week.append('Friday')
    if i==5:
        week.append('Saturday')
    if i==6:
        week.append('Sunday')
        
ac['weekday']=week
```

```python
time=[]
for i in ac['Hour']:
        if i in [22, 23, 0, 1, 2, 3]:
            time.append('Night')
        elif i in range(4, 12):
            time.append('Morning')
        elif i in range(12, 17):
            time.append('Afternoon')
        elif i in range(17, 22):
            time.append('Evening')
        
ac['hour']=time
```

```python
#total power consumed per minute
ac['Total']=ac[['AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7',
       'AC 8', 'AC 9', 'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15',
       'AC 16', 'AC 17', 'AC 18']].sum(axis=1)
```

```python
#to merge ac data and the temp data we need to make similar columns so as to merge on that 
ac['Year']=ac['Year'].astype(str)
ac['Month']=ac['Month'].astype(str)
ac['Day']=ac['Day'].astype(str)
ac['Hour']=ac['Hour'].astype(str)
```

```python
#merging both the dataframes on date column
final_df=pd.merge(ac,temp,on='date',how='inner')
```

```python
#dropping the other columns of  temp dataframe from final_df except temperature column
final_df.drop(['datetime', 'year', 'month_y', 'day', 'hour_y', 'weekday_y','Month_y', 'Weekday_y', 'Hour_y','date'],axis=1,inplace=True)
```

```python
final_df.columns
```

```python
#renaming the columns
final_df=final_df.rename(columns={'Month_x':'Month','Hour_x':'Hour','Weekday_x':'Weekday','month_x':'Month Name','weekday_x':'Weekday Name',
                         'hour_x':'Timing'})
```

```python
final_df.head()
```

```python
#forcasting the power consumption for each ac
a=['AC 1','AC 2','AC 3','AC 4','AC 5','AC 6','AC 7',
   'AC 8','AC 9','AC 10','AC 11','AC 12','AC 13',
   'AC 14','AC 15','AC 16','AC 17','AC 18']

final_df.groupby('temp')[a].mean().plot(figsize=(20,10))
plt.xlabel('Temperature')
plt.ylabel('AC')
plt.title("Forecasts for power consumption")
```
